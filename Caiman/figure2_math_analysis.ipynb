```python
import numpy as np
from sklearn.linear_model import LinearRegression


def regression_line(X,coef, intercept):
  y = intercept + coef * X
  return y

```

```python
########### 30 minutes ###########

# we extract the points
x30 = np.array(yoda1_1um_30min['dist_center']).reshape(-1,1)
y30 = yoda1_1um_30min['frequency']
# fit the model to a linear regression
model30 = LinearRegression().fit(x30, y30)
r_sq30 = model30.score(x30, y30)

r_square30 = model30.score(x30,y30)
# generate the F test
X30 = sm.add_constant(x30)
results30 = sm.OLS(y30, X30).fit()
A30 = np.identity(len(results30.params))
A30 = A30[1:,:]
# extract p value of the f_test
p_value30 = str(results30.f_test(A30)).split(',')[1].strip(' ')

# plot the figure
fig = plt.figure(figsize = (10,10), dpi = 600)

ax = fig.add_subplot(1,1,1)

ax.scatter(x30,y30, color = colors[0], label = '30 minutes')

ax.plot(regression_line(np.arange(250),coef = model30.coef_, intercept = model30.intercept_), color = colors[0], label = f'{p_value30}')

############ 60 min ###################
# we extract the values
x60 = np.array(yoda1_1um_60min['dist_center']).reshape(-1,1)
y60 = yoda1_1um_60min['frequency']

# fit the linear regression model
model60 = LinearRegression().fit(x60, y60)
r_sq60 = model60.score(x60, y60)

# generate the F test
X60 = sm.add_constant(x60)
results60 = sm.OLS(y60, X60).fit()
A60 = np.identity(len(results60.params))
A60 = A60[1:,:]
# extract p value of the f_test
p_value60 = str(results60.f_test(A60)).split(',')[1].strip(' ')
r_square60 = model60.score(x60,y60)
# plot the results

ax.scatter(x60,y60, color = colors[1], label = '60 minutes')

ax.plot(regression_line(np.arange(250),coef = model60.coef_, intercept = model60.intercept_), color = colors[1], label = f'{p_value60}')

########## 90 minutes ##############
#
# extract the valujes
#x90 = np.array(yoda1_1um_90min['dist_center']).reshape(-1,1)
#y90 = yoda1_1um_90min['frequency']

#model90 = LinearRegression().fit(x90, y90)
#r_sq90 = model90.score(x90, y90)

# generate the F test

#X90 = sm.add_constant(x90)
#results90 = sm.OLS(y90, X90).fit()
#A90 = np.identity(len(results90.params))
#A90 = A90[1:,:]
# extract p value of the f_test
#p_value90 = str(results90.f_test(A90)).split(',')[1].strip(' ')
#r_square90 = model90.score(x90,y90)
# plot the results

#ax.scatter(x90,y90, color = colors[2], label = '90 minutes')

#ax.plot(regression_line(np.arange(250),coef = model90.coef_, intercept = model90.intercept_), color = colors[2], label = f'{p_value90}')

ax.set_xlabel('Distance from the center (pixels)', fontsize= 12)
ax.set_ylabel('Frequency of oscilation of $Ca^{+2}$ (peaks/second)', fontsize = 12)
ax.legend(loc = 'upper right')
ax.grid()
ax.set_axisbelow(True)
plt.savefig('/content/drive/MyDrive/Internship/colabs/Images/dist_center_pvalue.svg')
```

```python
def regression_plane(x1, x2, intercept, coef1, coef2):

  z = intercept + coef1 * x1 + coef2 * x2

  return z
```
